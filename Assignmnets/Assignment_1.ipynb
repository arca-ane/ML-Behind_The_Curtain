 Q6 Compute the inverse and primary diagnol sum of x=[[1,2],[3,4]]

```python
x=np.array([[1,2],[3,4]])
print("Given Matrix : \n",x)
xinv=np.linalg.inv(x)
print("Inverse of Matrix : \n",xinv)
xdiagsum=np.trace(x)
print("Primary Diagonal sum :",xdiagsum)
```

Q7 Sort x along the second axis. x = [[1,4],[3,1]]

```python
x=np.array([[1,4],[3,1]])
xsort=np.sort(x,axis=1)
print("Original Matrix : \n",x)
print("Sorted along Second axis : \n",xsort)
```

Q8 Get the indices that would sort x(as above) along the second axis.

```python
x=np.array([[1,4],[3,1]])
xsort=np.argsort(x,axis=1)
print("Original Matrix : \n",x)
print("Sorted indices along Second axis : \n",xsort)
```

Run the following load command to get the 2-Dim iris dataset. Do the following operations on it.

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
```

Drop rows that contain a missing value. 

```python
droprow=iris_2d[~np.isnan(iris_2d).any(axis=1)]
#print(droprow)
print(np.size(droprow))
print(np.size(iris_2d))
```

Find the correlation between first two columns.

```python
col1,col2=iris_2d[:,1],iris_2d[:,2]
corr=np.corrcoef(col1,col2)
print(corr)
```

Find the mean,median and standard deviation of column 3.

```python
col3=iris_2d[:,3]
print("Mean is : ",np.mean(col3))
print("Median is : ",np.median(col3))
print("Standard Deviation is : ",np.std(col3))
```
